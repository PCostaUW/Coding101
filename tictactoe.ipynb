```python
#setup board
# Setup screen
screen = turtle.Screen()
screen.title("Tic Tac Toe")
screen.setup(width=400, height=400)


#global state
board = [["" for i in range(3)] for i in range(3)]
turn = 'X'
game_over = False

#Create a turtle for drawing
pen = turtle.Turtle()
pen.hideturtle()
pen.pensize(3)
pen.speed(0)

def draw_board():
    #two vertical lines
    pen.penup()
    pen.goto(-50,150)
    pen.pendown()
    pen.goto(-50,-150)

    pen.penup()
    pen.goto(50, 150)
    pen.pendown()
    pen.goto(50,-150)

    #two horizontal lines
    pen.penup()
    pen.goto(-150,50)
    pen.pendown()
    pen.goto(150,50)

    pen.penup()
    pen.goto(-150,-50)
    pen.pendown()
    pen.goto(150,-50)


def draw_symbol(row, col, symbol):
    #draws an X or O
    x = -100 + col*100
    y = 100 - row*100

    pen.penup()
    pen.goto(x,y-50)
    pen.pendown()
    pen.write(symbol, align='center', font=('Arial', 36, "normal"))


def check_winner():
    #check rows, columns, and diagonals
    for i in range(3):
        
        if board[i][0] == board[i][1] == board[i][2] != "":
            return board[i][0] #checks rows
            
        if board[0][i] == board[1][i] == board[2][i] != "":
            return board[0][i] #checks columns
            
    if board[0][0] == board[1][1] == board[2][2] != "":
        return board[0][0] #checks \ diag
        
    if board[0][2] == board[1][1] == board[2][0] != "":
        return board[0][2] #checks / diag
    
    return None #doesn't return anything if nobody has won

def is_full():
    #go through the rows of the board. If there are no empty spaces, return that the board is full
    for row in board:
        if "" in row:
            return False
    return True

def click(x,y):
    global turn, game_over

    if game_over:
        return

    col = int((x+150) // 100)
    row = int((-y + 150) // 100)

    if 0 <= row < 3 and 0 <= col < 3:
        if board[row][col] == "":
            board[row][col] = turn
            draw_symbol(row, col, turn)

            winner = check_winner()
            if winner:
                pen.penup()
                pen.goto(0, 160)
                pen.color("green")
                pen.write(f"{winner} wins!", align="center", font=("Arial", 24, "bold"))
                game_over = True
            elif is_full():
                pen.penup()
                pen.goto(0, 160)
                pen.color("blue")
                pen.write("It's a tie!", align="center", font=("Arial", 24, "bold"))
                game_over = True
            else:
                turn = "O" if turn == "X" else "X"


# Set up the game
draw_board()
screen.onclick(click)
screen.mainloop()
```
